# Brain Tumor MRI — Exploratory Analysis

Explore dataset structure, class balance, and preprocessing outputs.

Run `python scripts/download_data.py` before using this notebook.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import cv2

from src.data.dataset import count_samples_by_class
from src.data.preprocessing import PreprocessingPipeline
from src.utils.config import PreprocessingConfig, load_config

config = load_config("configs/base.yaml")
train_counts = count_samples_by_class(config.data.train_dir, config.data.class_names)
test_counts = count_samples_by_class(config.data.test_dir, config.data.class_names)
print("Train:", train_counts, "Total:", sum(train_counts.values()))
print("Test:", test_counts, "Total:", sum(test_counts.values()))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].bar(train_counts.keys(), train_counts.values())
ax[0].set_title("Train class counts")
ax[1].bar(test_counts.keys(), test_counts.values())
ax[1].set_title("Test class counts")
plt.tight_layout()
plt.show()
```

```python
pipeline = PreprocessingPipeline(config.preprocessing, config.data.image_size)
sample_dir = Path(config.data.train_dir) / "glioma"
sample_path = next(sample_dir.glob("*.jpg"))
raw = cv2.imread(str(sample_path))
processed = pipeline(raw)

fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].imshow(cv2.cvtColor(raw, cv2.COLOR_BGR2RGB))
ax[0].set_title("Raw")
ax[0].axis("off")
ax[1].imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))
ax[1].set_title("Preprocessed")
ax[1].axis("off")
plt.show()
```
